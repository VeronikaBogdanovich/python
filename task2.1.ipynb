```python
def find(arr):
    A = 0
    O = 0
    E = 0
    I = 0
    indexA = []
    indexO = []
    indexE = []
    indexI = []
    for i in range (len(arr)):
        if (arr[i]== 'а'):
            A += 1
            indexA.append(i)
        elif (arr[i]== 'о'):
            O += 1
            indexO.append(i) 
        elif (arr[i]== 'е'):
            E += 1
            indexE.append(i)
        elif (arr[i]== 'и'):
            I += 1
            indexI.append(i)
    if(A>0):
        print('а встречается:', A, 'раз, под индексами:', ' '.join(map(str,indexA)))
    if(O>0):
        print('о встречается:', O, 'раз, под индексами:', ' '.join(map(str,indexO)))
    if(E>0):
        print('е встречается:', E, 'раз, под индексами:', ' '.join(map(str,indexE)))
    if(I>0):
        print('и встречается:', I, 'раз, под индексами:', ' '.join(map(str,indexI)))
```

```python
def getPatronymic(arr):
    pytr = []
    for i in range (len(arr)):
        if arr[i] == ' ':
            k=i+1
            for k in range (len(arr)):
                if arr[k] == ' ':
                    pytr=arr[(k+1):]
    return len(pytr)
            
def delSpaces(arr):
    for i in range (len(arr)-2):
        if (arr[i] == ' '):
            del arr[i]
```

```python
def oneAndZero(let, glas):
    arr = []
    for k in range (len(let)):
        arr.append(let[k])
    for i in range (len(arr)):
        for j in range (len(glas)):
            if(arr[i] == glas[j]):
                arr[i] = 1
        if(arr[i] != 1):
            arr[i] = 0
    return arr

def split_numbers(string):
    splitstring=[]
    temp=string[:len(string)]
    for i in range(len(string)-1,0,-1):
        if 9<int(''.join(map(str,string[i:])),2) and int(''.join(map(str,string[i:])),2)<=100 :
            t=''.join(map(str,string[i:]))
            temp=''.join(map(str,string[:i]))
            break
    j=1
    n=len(temp)
    while j<n:
        if int(temp[:j],2)<=9 and int(temp[:j],2)>0:
            splitstring.append(''.join(map(str,temp[:j])))
            temp=temp[j:]
            j=0
            n=len(temp)
        if j==0 and temp==[0]*len(temp):
            temp=''
        j=j+1
    splitstring.append(t)
    return ' '.join(map(str,splitstring))

def makeNew (rus, eng):
    arr= []
    for i in range (len(eng)):
        rus.append(eng[i])
    for n in range (len(rus)):
        arr.append(rus[n])
    return arr
    
def avtoChange (rus, eng, avto):
    arrAvto = makeNew (rus,eng)
    for j in range (len(arrAvto)):
        for k in range (len(avto)):
            if (arrAvto[j] == avto[k]):
                arrAvto[j] = 2
        for l  in range (len(avto)):
            if (arrAvto[j] != 2 and arrAvto[j] == chr(ord('a') + l)):
                arrAvto[j] = 3
        if (arrAvto[j] != 2 and arrAvto[j] != 3):
            arrAvto[j] = 1
    return arrAvto
    
def getRusEngPatr(rus, eng, rusPatrLen, engPatrLen, lenEng):
    concatedWord = makeNew(rus,eng)
    engPatr = concatedWord[(len(concatedWord) - engPatrLen):]
    newWord = concatedWord[:(len(concatedWord) - (lenEng*2))]
    rusPatr = newWord[(len(newWord)-rusPatrLen):]
    for i in range (len(engPatr)):
        rusPatr.append(engPatr[i])
    print(''.join(map(str,rusPatr)))
    
    engNS = concatedWord[(len(concatedWord) - lenEng):(len(concatedWord) - engPatrLen)]
    rusNS = concatedWord[:(len(concatedWord) - (lenEng*2) - rusPatrLen)]
    rusToAppend = newString(rusNS)
    engToAppend = newString(engNS)
    
    string = []
    for j in range (len(engToAppend)):
        rusToAppend.append(engToAppend[j]) 
    for k in range (len(rusPatr)):
        string.append(rusPatr[k])
    for n in range (len(rusToAppend)):
        string.append(rusToAppend[n])
    print(''.join(map(str,string)))


def newString(arr): 
    letters = []
    for i in range (len(arr)):
        if (arr.count(arr[i])) == 2 or (arr.count(arr[i])) == 1:
            letters.append(arr[i])
    return letters    
```
